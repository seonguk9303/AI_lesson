```python
# udon 폴더를 WD로 지정
# 이미지 파일들의 크기를 128,128로 지정
# 32, 32, 96, 96 지점을 cropping을 해서 이미지를 처리
# 새로운 파일들은 udon_new라는 새로운 폴더에 저장
import os
from PIL import Image


os.getcwd()
if not os.path.isdir('./../udon_new'):
  os.makedirs('./../udon_new')
os.chdir("/content/udon")
files=os.listdir()
for file in files:
  if file.endswith(".png") | file.endswith(".jpg"): 
    temp = Image.open(file)
    temp2 = temp.resize( (128, 128) )
    temp2.save("resize_"+temp.filename)
    croppedIm = temp2.crop( (32,32,96,96))
    croppedIm.save("./../udon_new/cropping"+temp.filename)
```
